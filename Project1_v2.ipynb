```python
productinfotemp=[]
for url in productlink3:
    pages = np.arange(1, 13, 1)
    for page in pages:
        opt = webdriver.ChromeOptions()
        opt.add_argument("--disable-notifications")
        opt.add_argument("--disable-popup-blocking")
        opt.add_argument("start-maximized")
        opt.add_argument('disable-infobars')
        page = url + '/opt/page:' + str(page) + '/'
        path = r'C:\\Users\\tanisha16\\Downloads\\Chromedriver'
        driver = webdriver.Chrome(executable_path = path,options=opt)
        driver.get(page)
        time.sleep(3)
        try:
            driver.find_element_by_id("consent_prompt_submit").click()
            elems = driver.find_elements_by_tag_name('a')
            for elem in elems:
                try:
                    if 'product' in elem.get_attribute('href'):
                        productinfotemp.append(elem.get_attribute('href'))
                    else:    
                        productinfotemp.append("")
                except:
                        productinfotemp.append("")
        except:
            time.sleep(5)
            driver.find_element_by_id("consent_prompt_submit").click()
            elems = driver.find_elements_by_tag_name('a')
            for elem in elems:
                try:
                    if 'product' in elem.get_attribute('href'):
                        productinfotemp.append(elem.get_attribute('href'))
                    else:    
                        productinfotemp.append("")
                except:
                        productinfotemp.append("")
            
        driver.close()
```

### Capturing the Product Specific Info from Each Product Link

### Data Manipulation & conversion of the list to a readable dataframe

```python
import bs4
import requests

product = []
def extract_source(url):
        agent = {"User-Agent":"Mozilla/5.0"}
        source=requests.get(url, headers=agent).text
        return source
counter=0
for url in productinfo2:
        def extract_data(source):
                soup=bs4.BeautifulSoup(source, 'lxml')
                header=soup.findAll('h2')
                for i in header:
                    product.append(i.text)
                content=soup.findAll('li')
                for i in content:
                    product.append(i.text)
        extract_data(extract_source(url))
        counter+=1
        print("Current Website:",counter)
```

```python
product
```

```python
product = list(filter(None, product))

from itertools import groupby
def split_condition(x):
    return x in {'About this product'}
grouper = groupby(product, key=split_condition)
res = dict(enumerate((list(j) for i, j in grouper if not i), 1))
print(res)

df=pd.DataFrame.from_dict(res,orient='index')

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 1000)
df

df.shape

df1 = df.iloc[1:,14:]
df1
df1['Price'] = df1.apply(lambda row: row.apply(lambda x: x if re.search("^[£]",str(x)) else None), axis=1).fillna(" ").agg(' '.join, axis=1).apply(lambda x: x.strip().split(" ")[0])
df1['EAN'] = df1.apply(lambda row: row.apply(lambda x: x if re.search("EAN",str(x)) else None), axis=1).fillna(" ").agg(''.join, axis=1)#.reset_index(drop=True)

df2 = df1[[14,15,16,17,18,'EAN','Price']]
df2['EAN'] = df2['EAN'].str.replace('.','')
df2['EAN'] = df2['EAN'].str.replace(' ','')
df2['Price'] = df2['Price'].str.replace('*','')
df2 = df2.rename(columns={14:"Category",
                          15:"Sub-Cateogry",
                          16:"Product Type",
                          17:'Product Line',
                          18:"Product Line Item",
                          "EAN":"EAN Code",
                          "Price":"Product Price"
                   }
            )
df2 = df2[["Category","Sub-Cateogry","Product Type",'Product Line',"Product Line Item","Product Price","EAN Code"]]
```

```python
df2['Product Line Item'] = df2['Product Line Item'].str.replace('View fabricdetail','')
df2['Product Line Item'] = df2['Product Line Item'].str.replace('Video','')
df2['Product Line Item'] = np.where(df2['Product Line Item'].str.startswith('.cxd'),'',df2['Product Line Item'])
df2['Product Price'] = np.where(df2['Product Price'].str.startswith('.c'),'',df2['Product Price'])

df2 = df2[df2['Product Price']!=''].reset_index(drop=True)

df2.loc[df2['Product Line Item'].astype(str).str.startswith("£"), 'Product Line Item'] = ""

df2['Product Line Item'] = np.where(df2['Product Line Item']=="",df2['Product Line'],df2['Product Line Item'])
df2['Product Line'] = np.where(df2['Product Line']==df2['Product Line Item'],"",df2['Product Line'])

df2
```
